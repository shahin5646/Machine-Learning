#### Creating Sine wave graph

```python
# Compute the x and y coordinates for points on a sine curve

x= np.arange(0,5*np.pi,0.2)
y = np.sin(x)

plt.title('Sin Wave')
plt.plot(x,y,'g')
```

```python
#Creating Sin and cosine subplot

x = np.arange(0,5*np.pi,0.1)
y_sinx = np.sin(x)
y_cosx = np.cos(x)

#Creating subplot consist of 2 rows and 1 column
plt.title("Sin Wave")
plt.subplot(2,1,1)
plt.plot(x,y_sinx,'g--')


plt.subplot(2,1,2)
plt.plot(x,y_cosx,'r')
plt.title('Cos Wave')

plt.show() #without this we can run and it will show the output
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
